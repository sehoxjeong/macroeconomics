```julia
using LinearAlgebra
using Plots, LaTeXStrings

# Set model parameters.
β = 0.96 # time discount factor
α = 0.4  # capital share in production function
A = 1.0  # productivity
δ = 1.0  # capital depreciation rate

knum = 30
kmin = 1e-7
kmax = 4.0

kgrid = collect(range(kmin, kmax, knum))

u(c) = log(c)
f(k) = A * k^α

# Compute consumption matrix.
c = zeros(knum, knum)

for j in 1:knum
    c[:, j] = f.(kgrid) .+ (1 - δ) * kgrid .- kgrid[j]
end
```

```julia
heatmap(c, xlabel=L"k'", ylabel=L"k")
```
